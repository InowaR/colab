```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([[1.0, 1.0, 1.0]])
Y = np.array([[1.0]]).T

for i in range(200):
  w1 = np.random.random((3,3))
  w2 = np.random.random((3,1))
  S1 = 1 / (1 + np.exp(-(np.dot(X, w1))))
  S2 = 1 / (1 + np.exp(-(np.dot(S1, w2))))
  S2_delta = (Y - S2) * (S2 * (1 - S2))
  S1_delta = S2_delta.dot(w2.T) * (S1 * (1 - S1))
  w1 += X.T.dot(S1_delta)
  w2 += S1.T.dot(S2_delta)
  # print(w1)
  # print(w1[0][0], 'red')
  # print(w1[0][1], 'green')
  # print(w1[0][2], 'blue')
  # print("+-----------------------------------+")
  plt.scatter(w1[0][0], i, color = 'red')
  plt.scatter(w1[0][1], i, color = 'green')
  plt.scatter(w1[0][2], i, color = 'blue')
```
