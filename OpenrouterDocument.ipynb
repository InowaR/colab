```python
from google.colab import userdata
key = userdata.get('OPENROUTER_API_KEY')
```

```python
from openai import OpenAI
import base64
from PIL import Image
from io import BytesIO
import os # Для получения API ключа из переменной окружения

# Получаем API ключ из переменной окружения.
# Рекомендуется не хардкодить ключ напрямую в коде.
# Убедитесь, что у вас установлена переменная окружения с вашим ключом,
# например, OPENROUTER_API_KEY.


client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=key,
)

# Функция для кодирования изображения в Base64
def encode_image_to_base64(image_path):
    """
    Кодирует изображение из файла в строку Base64.
    """
    try:
        with Image.open(image_path) as img:
            buffered = BytesIO()
            # Сохраняем изображение в буфер в формате PNG.
            # PNG универсален и хорошо подходит для изображений с текстом.
            img.save(buffered, format="PNG")
            return base64.b64encode(buffered.getvalue()).decode('utf-8')
    except FileNotFoundError:
        print(f"Ошибка: Файл '{image_path}' не найден.")
        return None
    except Exception as e:
        print(f"Ошибка при обработке изображения: {e}")
        return None

# Укажите путь к вашему изображению
# Предполагается, что 'image.jpg' находится в той же папке, что и скрипт
image_file_path = 'image.jpg'

# Кодируем изображение
encoded_image = encode_image_to_base64(image_file_path)

if encoded_image:
    completion = client.chat.completions.create(
      model="google/gemini-2.0-flash-exp:free", # Используйте модель, которая поддерживает мультимодальность
      messages=[
        {
          "role": "user",
          "content": [
            {
              "type": "text",
              "text": "Пожалуйста, напечатай всё содержимое текстового документа, который находится на этой фотографии."
            },
            {
              "type": "image_url",
              "image_url": {
                # Формат "data:image/png;base64,{encoded_image}" указывает тип данных и саму Base64 строку
                "url": f"data:image/png;base64,{encoded_image}"
              }
            }
          ],
        }
      ]
    )
    print(completion.choices[0].message.content)
else:
    print("Не удалось отправить запрос, так как изображение не было закодировано.")
```
