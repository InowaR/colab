```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Определяем параметры
T_env = 20  # Температура окружающей среды (°C)
T0 = 100    # Начальная температура тела (°C)
k = 0.1     # Коэффициент охлаждения (1/мин)

# Определяем функцию, описывающую дифференциальное уравнение
def model(T, t):
    dTdt = -k * (T - T_env)
    return dTdt

# Время от 0 до 60 минут
t = np.linspace(0, 60, 100)

# Решаем дифференциальное уравнение
T = odeint(model, T0, t)

# Визуализируем результат
plt.plot(t, T, label='Температура тела')
plt.axhline(y=T_env, color='r', linestyle='--', label='Температура окружающей среды')
plt.title("Процесс охлаждения тела")
plt.xlabel("Время (минуты)")
plt.ylabel("Температура (°C)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем параметры
T_env = 20  # Температура окружающей среды (°C)
T0 = 100    # Начальная температура тела (°C)
k = 0.1     # Коэффициент охлаждения (1/мин)

# Время от 0 до 60 минут
t = np.linspace(0, 60, 100)

# Аналитическое решение
T_analytical = T_env + (T0 - T_env) * np.exp(-k * t)

# Визуализируем результат
plt.plot(t, T_analytical, label='Температура тела (аналитически)')
plt.axhline(y=T_env, color='r', linestyle='--', label='Температура окружающей среды')
plt.title("Процесс охлаждения тела")
plt.xlabel("Время (минуты)")
plt.ylabel("Температура (°C)")
plt.legend()
plt.grid()
plt.show()
```
