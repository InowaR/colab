```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
T_1 = 37  # начальная температура
T_0 = 0  # конечная температура
k = 0.01  # коэффициент теплоотдачи
T_control = 20 # температура 20 градусов

t_span = np.linspace(0, 1000, 1000) # время (начало, время, количество точек)

T = np.zeros_like(t_span) # температура
T[0] = T_1


for i in range(1, len(t_span)):
  T[i] = T_0 + (T_1 - T_0) * np.exp(-k * t_span[i-1])


t_control = t_span[np.where(T <= T_control)[0]][0] # время, при котором температура равна T_control
t_control_index = np.where(t_span == t_control)


plt.plot(t_span, T)
plt.xlabel("Время (с)")
plt.ylabel("Температура (°C)")


plt.scatter(t_control, T[t_control_index], color="red")
plt.axvline(x=t_control, color="red", linestyle="dashed")

plt.show()
```
