```python
from flask import Flask
from google.colab.output import eval_js

public_url = eval_js("google.colab.kernel.proxyPort(5000)")
print(public_url)


app = Flask(__name__)

@app.route('/')
def hello():
    return "Hello"

app.run(host='0.0.0.0', port=5000)
```
