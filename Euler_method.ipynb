```python
import matplotlib.pyplot as plt

"""
Программа для численного моделирования гармонического осциллятора
(пружинный маятник) методом Эйлера.
"""

# Физические параметры системы
M = 10   # Масса груза (кг)
K = 10   # Жесткость пружины (Н/м)

# Начальные условия системы
Xs = [1]  # Начальное положение: груз смещен на 1 м от положения равновесия
Vs = [0]  # Начальная скорость: груз отпускают без начальной скорости

# Параметры численного моделирования
dt = 0.04  # Шаг по времени (секунды) - чем меньше, тем точнее расчет
T = 10     # Общее время моделирования (секунды)

# Главный цикл расчета - численное интегрирование методом Эйлера
# Количество итераций = общее время / шаг времени
for step in range(int(T/dt)):
    # ШАГ 1: Расчет производных (скоростей изменения)
    # Скорость изменения положения - это текущая скорость
    dx_dt = Vs[-1]

    # Скорость изменения скорости - это ускорение
    # По второму закону Ньютона: F = m*a
    # Сила пружины: F = -k*x (направлена к положению равновесия)
    # Поэтому ускорение: a = F/m = (-k*x)/m
    dv_dt = -Xs[-1] * K / M

    # ШАГ 2: Интегрирование - нахождение новых значений
    # Новое положение = старое положение + изменение положения за время dt
    new_x = Xs[-1] + dx_dt * dt

    # Новая скорость = старая скорость + изменение скорости за время dt
    new_v = Vs[-1] + dv_dt * dt

    # Добавляем новые значения в массивы
    Xs.append(new_x)
    Vs.append(new_v)

# Визуализация результатов
plt.figure(figsize=(12, 6))

# Строим график скорости (красная линия)
plt.plot(Vs, color='red', linewidth=2, label=f'Скорость (м/с)')

# Строим график положения (синяя линия)
plt.plot(Xs, color='blue', linewidth=2, label=f'Положение (м)')

# Настраиваем внешний вид графика
plt.title('Гармонические колебания пружинного маятника\nМетод Эйлера', fontsize=14)
plt.xlabel('Шаг времени', fontsize=12)
plt.ylabel('Значение', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Отображаем график
plt.tight_layout()
plt.show()

# Выводим основную информацию о расчете
print("="*50)
print("РЕЗУЛЬТАТЫ МОДЕЛИРОВАНИЯ")
print("="*50)
print(f"Масса груза M: {M} кг")
print(f"Жесткость пружины K: {K} Н/м")
print(f"Отношение K/M: {K/M}")
print(f"Теоретический период колебаний: {2 * 3.14159 * (M/K)**0.5:.2f} с")
print(f"Шаг времени dt: {dt} с")
print(f"Общее время моделирования: {T} с")
print(f"Количество шагов расчета: {len(Xs)}")
print(f"Амплитуда колебаний положения: от {min(Xs):.3f} до {max(Xs):.3f} м")
print(f"Амплитуда колебаний скорости: от {min(Vs):.3f} до {max(Vs):.3f} м/с")
```
