```python
import matplotlib.pyplot as plt

"""
     
( )  .
"""

#   
M = 10   #   ()
K = 10   #   (/)

#   
Xs = [1]  #  :    1    
Vs = [0]  #  :     

#   
dt = 0.04  #    () -  ,   
T = 10     #    ()

#    -    
#   =   /  
for step in range(int(T/dt)):
    #  1:   ( )
    #    -   
    dx_dt = Vs[-1]

    #    -  
    #    : F = m*a
    #  : F = -k*x (   )
    #  : a = F/m = (-k*x)/m
    dv_dt = -Xs[-1] * K / M

    #  2:  -   
    #   =   +     dt
    new_x = Xs[-1] + dx_dt * dt

    #   =   +     dt
    new_v = Vs[-1] + dv_dt * dt

    #     
    Xs.append(new_x)
    Vs.append(new_v)

#  
plt.figure(figsize=(12, 6))

#    ( )
plt.plot(Vs, color='red', linewidth=2, label=f' (/)')

#    ( )
plt.plot(Xs, color='blue', linewidth=2, label=f' ()')

#    
plt.title('   \n ', fontsize=14)
plt.xlabel(' ', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

#  
plt.tight_layout()
plt.show()

#     
print("="*50)
print(" ")
print("="*50)
print(f"  M: {M} ")
print(f"  K: {K} /")
print(f" K/M: {K/M}")
print(f"  : {2 * 3.14159 * (M/K)**0.5:.2f} ")
print(f"  dt: {dt} ")
print(f"  : {T} ")
print(f"  : {len(Xs)}")
print(f"  :  {min(Xs):.3f}  {max(Xs):.3f} ")
print(f"  :  {min(Vs):.3f}  {max(Vs):.3f} /")
```
