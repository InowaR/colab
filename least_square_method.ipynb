```python
import numpy as np
import matplotlib.pyplot as plt

# Пример данных
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])

# Количество точек
n = len(x)

# Вычисляем параметры a и b для уравнения y = ax + b
A = np.vstack([x, np.ones(n)]).T
a, b = np.linalg.lstsq(A, y, rcond=None)[0]

# Выводим результаты
print(f"Наклон (a): {a}")
print(f"Пересечение (b): {b}")

# Строим график
plt.scatter(x, y, color='blue', label='Данные')
plt.plot(x, a*x + b, color='red', label='Линия регрессии')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Метод наименьших квадратов')
plt.grid()
plt.show()
```
