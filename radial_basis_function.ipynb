```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

# Генерация данных для задачи XOR
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 0])  # Метки для XOR

# Обучение модели SVM с использованием ядра 'rbf'
model = svm.SVC(kernel='rbf', gamma='auto')
model.fit(X, y)

# Предсказание значений
predictions = model.predict(X)

# Вывод результатов
print("Предсказания:", predictions)

# Визуализация результатов
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z > 0, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title("SVM для задачи XOR")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Генерация случайных данных
X, y = datasets.make_moons(n_samples=100, noise=0.1, random_state=42)

# Разделение данных на тренировочные и тестовые
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Создание модели SVM с RBF-ядром
model = SVC(kernel='rbf', gamma='scale')  # gamma='scale' автоматически выбирает значение

# Обучение модели
model.fit(X_train, y_train)

# Предсказание на тестовых данных
y_pred = model.predict(X_test)

# Оценка производительности модели
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Визуализация результатов
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', marker='o', label='Train data')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', marker='x', label='Test data')

# Создание сетки для визуализации границ решения
xx, yy = np.meshgrid(np.linspace(-1.5, 2.5, 100), np.linspace(-1.0, 1.5, 100))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z > 0, alpha=0.3)
plt.title('SVM with RBF Kernel')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Загрузка набора данных Ирисы
iris = datasets.load_iris()
X = iris.data[:, :2]  # Используем только первые два признака для визуализации
y = iris.target

# Для простоты, выберем только два класса (0 и 1)
X = X[y != 2]
y = y[y != 2]

# Разделение данных на тренировочные и тестовые
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Создание модели SVM с RBF-ядром
model = SVC(kernel='rbf', gamma='scale')  # gamma='scale' автоматически выбирает значение

# Обучение модели
model.fit(X_train, y_train)

# Предсказание на тестовых данных
y_pred = model.predict(X_test)

# Оценка производительности модели
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Визуализация результатов
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', marker='o', label='Train data')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', marker='x', label='Test data')

# Создание сетки для визуализации границ решения
xx, yy = np.meshgrid(np.linspace(4, 8, 100), np.linspace(1.5, 5.5, 100))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z > 0, alpha=0.3)
plt.title('SVM with RBF Kernel on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from keras.datasets import mnist
from keras.utils import to_categorical

# Загрузка датасета MNIST
(X_train_full, y_train_full), (X_test, y_test) = mnist.load_data()

# Преобразование данных
X_train_full = X_train_full.reshape(-1, 28 * 28).astype('float32') / 255.0  # Нормализация
X_test = X_test.reshape(-1, 28 * 28).astype('float32') / 255.0  # Нормализация

# Для простоты, используем только первые 1000 образцов для обучения и тестирования
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.1, random_state=42)

# Создание модели SVM с RBF-ядром
model = SVC(kernel='rbf', gamma='scale', random_state=42)

# Обучение модели
model.fit(X_train[:1000], y_train[:1000])  # Используем только 1000 примеров для ускорения

# Предсказание на тестовых данных
y_pred = model.predict(X_val)

# Оценка производительности модели
print("Classification Report:")
print(classification_report(y_val, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred))

# Визуализация некоторых предсказаний
n = 10  # Количество изображений для отображения
plt.figure(figsize=(12, 4))
for i in range(n):
    plt.subplot(2, n, i + 1)
    plt.imshow(X_val[i].reshape(28, 28), cmap='gray')
    plt.title(f'Pred: {y_pred[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
