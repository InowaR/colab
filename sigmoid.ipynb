```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Значения x для вычисления
x_values = np.array([-10, 0, 1, 10])
# Вычисление значений сигмоиды для заданных x
sigmoid_values = sigmoid(x_values)

# Создание массива x для построения графика
x_plot = np.linspace(-10, 10, 100)
y_plot = sigmoid(x_plot)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='Сигмоида', color='blue')
plt.scatter(x_values, sigmoid_values, color='red', zorder=5)
plt.title('График функции сигмоиды')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-0.1, 1.1)

# Добавление аннотаций для отмеченных точек
for (x, y) in zip(x_values, sigmoid_values):
    plt.annotate(f'({x}, {y:.2f})', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Показать график
plt.show()
```
